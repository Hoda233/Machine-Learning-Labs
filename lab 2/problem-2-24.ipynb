### 2.24 (B)

```python
def target_function(x):
    # TODO: implement target function
    return x**2
```

```python
def run_numerical_experiment(size):
    # TODO: implement experiment

    bias = []
    variance = []
    e_out = []
    g_bar_list =[]

    x = np.random.uniform(-1, 1, size) 
    y = target_function(x)

    for xi in x:

        e_out2 = []
        g = []

        for _ in range(size):

            yi = target_function(xi)

            x1, x2 = np.random.uniform(-1, 1, 2)

            a = x1 + x2
            b = - x1 * x2

            gd = a * xi + b
            g.append(gd)

            e = (gd - yi)**2
            e_out2.append(e) 

        g_bar = np.mean(g)
        g_bar_list.append(g_bar)
        e_out.append(np.mean(e_out2))
        bias.append((g_bar - yi)**2)
        variance.append(np.mean((g - g_bar)**2)) 
    
    x = np.linspace(-1,1,1000)
    y = target_function(x)
    plt.plot(x,y, label = 'f(x)')
    plt.plot(x, [np.mean(g_bar_list) for _ in range(1000)], label = 'g_bar(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

    return np.mean(bias), np.mean(variance), np.mean(e_out)
```

### 2.24 (C)

```python
# TODO: Run experiment and calculate bias and variance (+ plotting)
bias, variance, eout = run_numerical_experiment(1000)
print(f"bias             {bias}")
print(f"variance         {variance}")
print(f"bias + variance  {bias + variance}")
print(f"Eout             {eout}")
```

### 2.24 (E)

```python
def run_numerical_experiment(size, d):
    # TODO: implement experiment

    bias = []
    variance = []
    e_out = []
    g_bar_list =[]

    x = np.random.uniform(-1, 1, (size, d)) 
    y = target_function(x)

    for xi in x:

        e_out2 = []
        g = []

        for _ in range(size):

            yi = target_function(xi)
            
            gd = 0
            for _ in range(d):
                x = np.random.uniform(-1, 1)
                gd += (x**2)
            gd/= d
            g.append(gd)

            e = (gd - yi)**2
            e_out2.append(e) 

        g_bar = np.mean(g)
        g_bar_list.append(g_bar)
        e_out.append(np.mean(e_out2))
        bias.append((g_bar - yi)**2)
        variance.append(np.mean((g - g_bar)**2)) 
    
    x = np.linspace(-1,1,1000)
    y = target_function(x)
    plt.plot(x,y, label = 'f(x)')
    plt.plot(x, [np.mean(g_bar_list) for _ in range(1000)], label = 'g_bar(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

    return np.mean(bias), np.mean(variance), np.mean(e_out)
```

```python
for d in [2, 5, 10, 20]:
    bias, variance, eout = run_numerical_experiment(1000,d)
    print(f"number of input data points {d}")
    print(f"bias             {bias}")
    print(f"variance         {variance}")
    print(f"bias + variance  {bias + variance}")
    print(f"Eout             {eout}")
```
